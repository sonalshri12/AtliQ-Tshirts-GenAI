```python
import sqlite3

# 1. DATABASE SETUP
conn = sqlite3.connect('atliq_tshirts.db')
cursor = conn.cursor()
cursor.execute('DROP TABLE IF EXISTS t_shirts')
cursor.execute('''CREATE TABLE t_shirts (
    id INTEGER PRIMARY KEY, brand TEXT, color TEXT, stock_quantity INTEGER
)''')
cursor.executemany('INSERT INTO t_shirts VALUES (?,?,?,?)', 
                   [(1, 'Nike', 'White', 50), (2, 'Adidas', 'Black', 20), (3, 'Nike', 'Red', 10)])
conn.commit()

# 2. LOGIC DEMONSTRATION (What the AI does)
def mock_ai_query(question):
    print(f"> Entering new SQLDatabaseChain chain...")
    print(f"Question: {question}")
    
    # This is the SQL the AI is trained to generate
    sql_query = "SELECT SUM(stock_quantity) FROM t_shirts WHERE brand = 'Nike';"
    print(f"AI Generated SQL: {sql_query}")
    
    cursor.execute(sql_query)
    result = cursor.fetchone()[0]
    print(f"Answer: {result}")
    print("> Finished chain.")

# 3. RUN
mock_ai_query("How many total Nike t-shirts are in stock?")
conn.close()
```
